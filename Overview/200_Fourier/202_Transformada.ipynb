```python
# Matemáticas - estadísticas
import numpy as np
import math

# Visualización
import seaborn as sns
import matplotlib.pyplot as plt

# Interactividad
import ipywidgets
```

```python
# Definir la función original a trozos
def original_function(x):
    return np.where((x > 0), 3, 0)

# Parámetros
frequencies = np.linspace(-10, 10, 1000)  # Frecuencias de -10 a 10
x_values = np.linspace(-5, 5, 1000)  # Valores de x de -2 a 2

# Definir la función de la transformada de Fourier
def fourier_transform(frequency):
    if frequency == 0:
        return 1  # Evitar la división por cero
    #return 3*np.sin(10*np.pi * frequency) / (10*np.pi * frequency)
    return (-3/(2*np.pi*frequency))*np.sin(-10*np.pi*frequency)
    #return (-3/(2*np.pi*(-1)**(1/2))*frequency) * (np.cos(-10*np.pi*frequency) - (-1)**(1/2)*np.sin(-10*np.pi*frequency)) + (-3/(2*np.pi*(-1)**(1/2))*frequency)
    
# Definir la función inversa de la transformada de Fourier
def inverse_fourier_transform(x, frequencies):
    result = 0
    for freq in frequencies:
        result += fourier_transform(freq) * np.exp(2j * np.pi * freq * x)
    return result

# Reconstruir la función original
reconstructed_function = np.real(inverse_fourier_transform(x_values, frequencies))

# Normalizar la reconstrucción
reconstructed_function *= np.sum(original_function(x_values)) / np.sum(reconstructed_function)

# Graficar la función original y la reconstruida
plt.plot(x_values, original_function(x_values), label='Función original')
plt.plot(x_values, reconstructed_function, label='Función reconstruida')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Reconstrucción de la función original')
plt.legend()
plt.show()
```

```python
an = integrate((1 / pi) * cos(n * t) * t**2, (t, 0, 2*pi ))

bn = integrate((1 / pi) * sin(n * t) * t**2, (t, 0, 2*pi ))

armonicos = 100
serie = ((an*cos(n*t)).subs(n, 0))/2
for i in range(1, armonicos + 1):
    serie = serie + (an*cos(n*t)).subs(n, i)
for j in range(1, armonicos + 1):
    serie = serie + (bn*sin(n*t)).subs(n, j)
plotting.plot(serie, ylim=(0, 41), xlim=(0,2*pi))
```
