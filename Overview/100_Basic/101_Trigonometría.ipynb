Las principales (hasta el momento creo que las únicas) funciones trigonométricas que se implementan para nuestro análisis objetivo son las correspondientes a ***seno*** y ***coseno***, estas funciones parten del análisis de un triángulo rectángulo para determinar diferentes medidas de distancias a una determinada circunferencia.

Ambas funciones trigonométricas (seno y coseno) reciben el rango $[-1,1]$ para cualquier número perteneciente al conjunto de los reales, la explicación de este intervalo se deduce a partir de la interpretación geométrica y visual de la misma, teniendo como referencia la imágen anterior, visualicemos el triángulo cuando el lado opuesto posee una longitud igual a cero, esto equivaldría a una recta en el plano cartesiano donde el ángulo $\theta$ equivale a cero y al reemplazar los valores en la formula correspondiente a la función ***seno*** obtemos el fraccionario $\frac{0}{r}$ donde el cociente equivale a cero. Realizando un análisis similar se logra inferir la lógica detrás de la función ***coseno***.

```python
# Crear espacio del dominio
time = np.linspace(-5,5,1000)
seno = np.sin(time)
coseno = np.cos(time)

# Vizualizar funciones
plt.plot(time,seno,'b-', time, coseno, 'y-')
plt.legend(['sen(x)','cos(x)'])
plt.axis([-5,5,-1.2,1.2])
plt.grid(True)
plt.title("Comparación de funciones trigonométricas")
plt.show()
```

# Aproximación mendiante series de Taylor

Cada función trigonométrica posee una aproximación polinomial llamada ***serie de Taylor y Maclaurin***, cada aproximación se difene como una sumatoria de polinomios con límite al infinito. La razón por la cuál el límite superior tiende a infinito se debe a que éste debe abarcar todo el dominio de la función seno (pese a ser periódica), en caso de solo considerar un límite superior específico, la ***serie de Taylor*** solo se aproxima a una parte del dominio de la ***función trigonométrica***.

La serie de Taylor que nos conviene es la correspondiente a la aproximación de la función trigonométrica ***seno***, se define de la siguiente manera.

$$\sum_{n=0}^{\infty} \frac{(-1)^{n}}{(2n+1)!} \dot{x^{2n+1}}$$

```python
def taylor_sin(limit=1, interval=5):
    """
    Función interactiva que permite observar las variaciones que posee
    la serie de Taylor en función de su límite superior y el intervalo
    del dominio, los argumentos de la función son respectivamente los
    anteriores mencionados.
    """

    # Crea el espacio time
    time = np.linspace(-interval,interval,1000)
    y = np.sin(time)
    approx_taylor = 0
    
    # Genero la serie de Taylor hasta el límite power
    for n in range(0, limit, 1):
        
        numerator = ((-1) ** (n)) * (time ** (2*n+1))
        denominator = math.factorial(2*n+1)
        approx_taylor += numerator / denominator

    # Generación de la gráfica
    plt.plot(time,y,'b-', time, approx_taylor, 'r--')
    plt.legend(['seno(x)','Taylor'])
    plt.axis([-interval,interval,-1.2,1.2]) # El rango se define en este intervalo para no ver la gráfica tan "apretada"
    plt.grid(True)
    plt.title(f"Aproximación hasta $x^{limit}$")
    plt.show()


ipywidgets.interact(taylor_sin, limit=(1,9,2), interval=(5,10,1))
```
