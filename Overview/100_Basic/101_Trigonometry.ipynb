Both trigonometric functions (sine and cosine) produce values within the range $[-1,1]$ for any number belonging to the set of real numbers. The explanation of this interval is derived from their geometric and visual interpretation. Referring to the image above, let us visualize the triangle when the opposite side has a length equal to zero. This scenario corresponds to a straight line in the Cartesian plane where the angle $\theta$ equals zero. Substituting these values into the formula for the ***sine*** function gives the fraction $\frac{0}{r}$, where the quotient equals zero. By conducting a similar analysis, we can deduce the logic behind the ***cosine*** function.

```python
# Crear espacio del dominio
time = np.linspace(-5,5,1000)
seno = np.sin(time)
coseno = np.cos(time)

# Vizualizar funciones
plt.plot(time,seno,'b-', time, coseno, 'y-')
plt.legend(['sen(x)','cos(x)'])
plt.axis([-5,5,-1.2,1.2])
plt.grid(True)
plt.title("Comparación de funciones trigonométricas")
plt.show()
```

# Approximation using Taylor Series

Each trigonometric function has a polynomial approximation called the ***Taylor and Maclaurin series***. Each approximation is defined as a summation of polynomials with an upper limit extending to infinity. The reason the upper limit tends to infinity is that it must cover the entire domain of the sine function (despite being periodic). If only a specific upper limit is considered, the ***Taylor series*** will only approximate a part of the domain of the ***trigonometric function***.

The Taylor series that is relevant to us is the one corresponding to the approximation of the ***sine*** trigonometric function. It is defined as follows:

$$\sum_{n=0}^{\infty} \frac{(-1)^{n}}{(2n+1)!} \dot{x^{2n+1}}$$

```python
# Define the limits and intervals for the 4 graphs
limits = [3, 5, 7, 9]
intervals = [5, 6, 8, 10]

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Iterate over the axes and plot each graph
for i, ax in enumerate(axs.flatten()):
    limit = limits[i]
    interval = intervals[i]
    
    # Create time values
    time = np.linspace(-interval, interval, 1000)
    y = np.sin(time)
    approx_taylor = 0
    
    # Generate the Taylor series approximation
    for n in range(0, limit, 1):
        numerator = ((-1) ** (n)) * (time ** (2*n+1))
        denominator = math.factorial(2*n+1)
        approx_taylor += numerator / denominator
    
    # Plot on the current axis
    ax.plot(time, y, 'b-', label='sin(x)')
    ax.plot(time, approx_taylor, 'r--', label=f'Taylor approx. $x^{limit}$')
    ax.set_xlim([-interval, interval])
    ax.set_ylim([-1.2, 1.2])
    ax.grid(True)
    ax.set_title(f"Approximation up to $x^{limit}$")
    ax.legend()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
